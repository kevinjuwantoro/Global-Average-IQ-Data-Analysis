### **a). Adding IQ Range column**

```python

# Define the IQ ranges
iq_bins = [float('-inf'), 20, 35, 50, 70, 85, 90, 110, 120, 130, 145, 160, 180, float('inf')]
iq_labels = ['<20', '20-34', '35-49', '50-69', '70-84', '85-89', '90-109', '110-119', '120-129', '130-144', '145-159', '160-179', '180+']

# Add a new column 'IQ Range'
df['IQ Range'] = pd.cut(df['Average IQ'], bins=iq_bins, labels=iq_labels, include_lowest=True)

# Display the updated DataFrame
df.head()
```

### **b). Adding IQ Classification column**

```python
# Define the IQ ranges and corresponding classifications
iq_ranges = [
    (0, 20, 'Profound intellectual disability'),
    (20, 35, 'Severe intellectual disability'),
    (35, 50, 'Moderate intellectual disability'),
    (50, 70, 'Mild intellectual disability'),
    (70, 85, 'Borderline intellectual functioning'),
    (85, 90, 'Below average'),
    (90, 110, 'Average intelligence'),
    (110, 120, 'Above average'),
    (120, 130, 'Moderately gifted'),
    (130, 145, 'Gifted'),
    (145, 160, 'Very gifted'),
    (160, 180, 'Almost genius'),
    (180, 200, 'Genius'),  
]

# Function to determine IQ range classification
def get_iq_classification(iq):
    for start, end, classification in iq_ranges:
        if start <= iq < end:
            return classification

# Add the 'Classification' column based on IQ range
df['Classification'] = df['Average IQ'].apply(lambda x: get_iq_classification(x))

# Display the updated DataFrame
df.head()
```

## **Filling The Missing value in HDI 2021**

```python
# Fill missing 'HDI (2021)' values with 'Literacy Rate' where 'HDI (2021)' is missing

df['HDI (2021)'] = df.groupby('Literacy Rate')['HDI (2021)'].transform(lambda x: x.fillna(x.median()))
df.head()
```

**The reason for filling the missing values of ‘HDI (2021)’ by grouping by ‘Literacy Rate’ is based on the assumption that countries with similar literacy rates might have similar Human Development Index (HDI) values.**

1. The HDI is a summary measure of average achievement in key dimensions of human development: a long and healthy life, being knowledgeable (hence the literacy rate), and have a decent standard of living. Therefore, it’s reasonable to assume that countries with similar literacy rates might have similar levels of human development, and hence, similar HDI values.

2. By using the median HDI of each literacy rate group to fill the missing HDI values, we’re making an educated guess that the missing HDI value would be close to the median of other countries with a similar literacy rate. The median is used instead of the mean to minimize the impact of any outliers in HDI within each literacy rate group.

However, it’s important to note that while this method can be a good starting point, it’s a simplification. There are many factors that contribute to a country’s HDI beyond the literacy rate, and this method doesn’t account for those.

## **Filling The Missing value in Mean Years of schooling - 2021**

```python
# Fill missing 'Mean years of schooling - 2021' values based on the mode within each 'Continent' group
df['Mean years of schooling - 2021'] = df.groupby('Continent')['Mean years of schooling - 2021'].transform(lambda x: x.fillna(x.mode().iloc[0]))
df.head()
```


**The reason for filling the missing values of ‘Mean years of schooling - 2021’ by grouping by ‘Continent’ is based on the assumption that countries within the same continent might have similar educational systems, policies, and cultural values towards education, which could result in similar ‘Mean years of schooling’.**

1. The ‘Mean years of schooling’ is a measure of the average number of years of education received by people ages 25 and older in their lifetime based on education attainment levels of the population converted into years of schooling based on theoretical duration of each level of education attended. Countries within the same continent often have similar socio-economic development and educational policies, which could lead to similar ‘Mean years of schooling’.

2. By using the mode (the most frequently occurring value) of ‘Mean years of schooling - 2021’ of each continent group to fill the missing values, we’re making an educated guess that the missing ‘Mean years of schooling - 2021’ value would be close to the most common ‘Mean years of schooling - 2021’ of other countries within the same continent.

However, it’s important to note that while this method can be a good starting point, it’s a simplification. There are many factors that contribute to a country’s ‘Mean years of schooling’ beyond the continent it’s in, and this method doesn’t account for those.

```python
# Selecting numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Creating boxplots for numerical columns arranged in a 2x5 grid
fig, axes = plt.subplots(2, 5, figsize=(20, 10))

for i, column in enumerate(numerical_columns):
    sns.boxplot(x=df[column], ax=axes[i // 5, i % 5])
    axes[i // 5, i % 5].set_title(f'Boxplot of {column}')
    axes[i // 5, i % 5].set_xlabel(column)

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Dictionary to store outliers' information
outliers_info = {}

for column in numerical_columns:
    # Calculate Q1, Q3, and IQR
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Calculate bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers_condition = (df[column] < lower_bound) | (df[column] > upper_bound)
    outliers = df[outliers_condition]

    # Store information in the dictionary
    outliers_info[column] = {
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'number_of_outliers': outliers_condition.sum(),  # Count of outliers
        'outliers_indices': outliers.index.tolist()  # Indices of outliers
    }


# summary for outliers_summary

outliers_summary = {key: {'bounds': (value['lower_bound'], value['upper_bound']), 
                          'number_of_outliers': value['number_of_outliers']} 
                    for key, value in outliers_info.items()}

# Creating DataFrame from 'outliers_summary'
outliers_summary_df = pd.DataFrame.from_dict(outliers_summary, orient='index')

# Splitting 'bounds' into 'lower_bound' and 'upper_bound' columns
outliers_summary_df[['lower_bound', 'upper_bound']] = outliers_summary_df['bounds'].apply(pd.Series)

# Dropping the original 'bounds' column
outliers_summary_df.drop('bounds', axis=1, inplace=True)

# Reordering the columns for a better view
outliers_summary_df = outliers_summary_df[['lower_bound', 'upper_bound', 'number_of_outliers']]

# Display DataFrame as a styled table
styled_table = outliers_summary_df.style.format('{:.0f}')
styled_table
```

## **Hypothesis Testing Average IQ Vs Literacy Rates**

### **a. Hypothesis Testing Average IQ vs Literacy Rates Global Country**

```python
from scipy.stats import kruskal

# Assuming 'Literacy Category' and 'Average IQ' columns are already created in the DataFrame
# Replace 'Literacy Category' and 'Average IQ' with your actual column names

# Categories for literacy rate
low_literacy = df[df['Literacy Category'] == 'Low']['Average IQ']
moderate_literacy = df[df['Literacy Category'] == 'Moderate']['Average IQ']
high_literacy = df[df['Literacy Category'] == 'High']['Average IQ']

# Perform Kruskal-Wallis H-test for Average IQ across different literacy rate categories
iq_kw_stat, iq_kw_p_value = kruskal(low_literacy, moderate_literacy, high_literacy)

# Set significance level (alpha)
alpha = 0.05

# Null and Alternative Hypotheses
null_hypothesis = "H0: There is no significant difference in Average IQ among different literacy rate categories."
alt_hypothesis = "Ha: At least one literacy rate category has a significantly different Average IQ."

# Check if the p-value is less than alpha to determine significance
if iq_kw_p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion:", alt_hypothesis)
    print(f"Kruskal-Wallis H-test statistic: {iq_kw_stat}")
    print(f"P-value: {iq_kw_p_value}")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion:", null_hypothesis)
```

### **b. Hypothesis Testing Average IQ vs Literacy Rates Asia Country**

```python

# Assuming 'Literacy Category', 'Average IQ', and 'Continent' columns are available in your DataFrame

# Filter the DataFrame for Asia continent and the specified literacy rate categories
asia_df = df[df['Continent'] == 'Asia']
low_literacy_asia = asia_df[asia_df['Literacy Category'] == 'Low']['Average IQ']
moderate_literacy_asia = asia_df[asia_df['Literacy Category'] == 'Moderate']['Average IQ']
high_literacy_asia = asia_df[asia_df['Literacy Category'] == 'High']['Average IQ']

# Perform Kruskal-Wallis H-test for Average IQ across different literacy rate categories in Asia
iq_kw_stat, iq_kw_p_value = kruskal(low_literacy_asia, moderate_literacy_asia, high_literacy_asia)

# Set significance level (alpha)
alpha = 0.05

# Null and Alternative Hypotheses
null_hypothesis = "H0: There is no significant difference in Average IQ among different literacy rate categories within the Asia continent."
alt_hypothesis = "Ha: At least one literacy rate category has a significantly different Average IQ within the Asia continent."

# Check if the p-value is less than alpha to determine significance
if iq_kw_p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion:", alt_hypothesis)
    print(f"Kruskal-Wallis H-test statistic: {iq_kw_stat}")
    print(f"P-value: {iq_kw_p_value}")
else:
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion:", null_hypothesis)
```

## **Hypothesis Testing Average IQ Vs HDI 2021**


### **a. Hypothesis Testing Average IQ vs HDI 2021 Global Country**

```python

# Categories for HDI
low_hdi = df[df['HDI Category'] == 'Low']['Average IQ']
medium_hdi = df[df['HDI Category'] == 'Medium']['Average IQ']
high_hdi = df[df['HDI Category'] == 'High']['Average IQ']

# Perform Kruskal-Wallis H-test for Average IQ across different HDI categories
iq_kw_stat, iq_kw_p_value = kruskal(low_hdi, medium_hdi, high_hdi)

# Set significance level (alpha)
alpha = 0.05

# Null and Alternative Hypotheses
null_hypothesis = "H0: There is no significant difference in Average IQ among different HDI categories."
alt_hypothesis = "Ha: At least one HDI category has a significantly different Average IQ."

# Check if the p-value is less than alpha to determine significance
if iq_kw_p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion:", alt_hypothesis)
    print(f"Kruskal-Wallis H-test statistic: {iq_kw_stat}")
    print(f"P-value: {iq_kw_p_value}")
else:
    print(f"Kruskal-Wallis H-test statistic: {iq_kw_stat}")
    print(f"P-value: {iq_kw_p_value}")
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion:", null_hypothesis)
```

### **b. Hypothesis Testing Average IQ vs HDI 2021 Asia Country**

```python
# Group HDI classifications for statistical analysis within Asia
low_hdi_asia = df[(df['Continent'] == 'Asia') & (df['HDI Category'] == 'Low')]['Average IQ']
medium_hdi_asia = df[(df['Continent'] == 'Asia') & (df['HDI Category'] == 'Medium')]['Average IQ']
high_hdi_asia = df[(df['Continent'] == 'Asia') & (df['HDI Category'] == 'High')]['Average IQ']

# Perform Kruskal-Wallis H-test for Average IQ across different HDI Categories in Asia
iq_kw_stat, iq_kw_p_value = kruskal(
    low_hdi_asia,
    medium_hdi_asia,
    high_hdi_asia
)

# Set significance level (alpha)
alpha = 0.05

# Null and Alternative Hypotheses
null_hypothesis = "H0: There is no significant difference in Average IQ among different HDI Categories in Asia."
alt_hypothesis = "Ha: At least one HDI Category has a significantly different Average IQ in Asia."

# Check if the p-value is less than alpha to determine significance
if iq_kw_p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion:", alt_hypothesis)
    print(f"Kruskal-Wallis H-test statistic: {iq_kw_stat}")
    print(f"P-value: {iq_kw_p_value}")
else:
    print(f"Kruskal-Wallis H-test statistic: {iq_kw_stat}")
    print(f"P-value: {iq_kw_p_value}")
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion:", null_hypothesis)

```

the null hypothesis (H0) states that there is no significant difference in Average IQ among different HDI (Human Development Index) Categories in Asia.
The test statistic is 3.74 and the p-value is 0.154. The p-value is greater than the common alpha level of 0.05, which means that the result is not statistically significant at the 5% level. Therefore, you fail to reject the null hypothesis.

This suggests that based on the data you have, there is not enough evidence to conclude that there is a significant difference in Average IQ among different HDI Categories in Asia. It’s important to note that failing to reject the null hypothesis is not the same as accepting it. It simply means that the data does not provide strong evidence against it.

However, this does not mean that there is absolutely no difference in Average IQ among different HDI Categories in Asia. There could be other factors not considered in the study that might influence the Average IQ. Also, the result might change with a larger sample size or different data.

Remember, statistical tests do not prove anything definitively. They simply provide evidence to support or refute a claim based on a particular dataset. It’s always a good idea to consider other evidence and use domain knowledge when interpreting the results of a statistical test.

## **Hypothesis Testing Average IQ Vs Population 2023**

### **a. Hypothesis Testing Average IQ vs Population 2023 Global Country**

```python
# Grouping Average IQ based on Population Classification
small_population = df[df['Population_Classification'] == 'Small']['Average IQ']
medium_population = df[df['Population_Classification'] == 'Medium']['Average IQ']
large_population = df[df['Population_Classification'] == 'Large']['Average IQ']
very_large_population = df[df['Population_Classification'] == 'Very Large']['Average IQ']

# Perform Kruskal-Wallis H-test for Average IQ across different Population Classifications
population_kw_stat, population_kw_p_value = kruskal(small_population, medium_population, large_population, very_large_population)

# Set significance level (alpha)
alpha = 0.05

# Null and Alternative Hypotheses
null_hypothesis = "H0: There is no significant difference in Average IQ among different Population Classifications."
alt_hypothesis = "Ha: At least one Population Classification has a significantly different Average IQ."

# Check if the p-value is less than alpha to determine significance
if population_kw_p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion:", alt_hypothesis)
    print(f"Kruskal-Wallis H-test statistic: {population_kw_stat}")
    print(f"P-value: {population_kw_p_value}")
else:
    print(f"Kruskal-Wallis H-test statistic: {population_kw_stat}")
    print(f"P-value: {population_kw_p_value}")
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion:", null_hypothesis)
```

### **b. Hypothesis Testing Average IQ vs Population 2023 in Asia Country**

```python
# Grouping Average IQ based on Population Classification within Asia
small_population_asia = df[(df['Continent'] == 'Asia') & (df['Population_Classification'] == 'Small')]['Average IQ']
medium_population_asia = df[(df['Continent'] == 'Asia') & (df['Population_Classification'] == 'Medium')]['Average IQ']
large_population_asia = df[(df['Continent'] == 'Asia') & (df['Population_Classification'] == 'Large')]['Average IQ']
very_large_population_asia = df[(df['Continent'] == 'Asia') & (df['Population_Classification'] == 'Very Large')]['Average IQ']

# Perform Kruskal-Wallis H-test for Average IQ across different Population Classifications in Asia
population_kw_stat, population_kw_p_value = kruskal(
    small_population_asia,
    medium_population_asia,
    large_population_asia,
    very_large_population_asia
)

# Set significance level (alpha)
alpha = 0.05

# Null and Alternative Hypotheses
null_hypothesis = "H0: There is no significant difference in Average IQ among different Population Classifications in Asia."
alt_hypothesis = "Ha: At least one Population Classification has a significantly different Average IQ in Asia."

# Check if the p-value is less than alpha to determine significance
if population_kw_p_value < alpha:
    print("Result: Reject the null hypothesis.")
    print("Conclusion:", alt_hypothesis)
    print(f"Kruskal-Wallis H-test statistic: {population_kw_stat}")
    print(f"P-value: {population_kw_p_value}")
else:
    print(f"Kruskal-Wallis H-test statistic: {population_kw_stat}")
    print(f"P-value: {population_kw_p_value}")
    print("Result: Fail to reject the null hypothesis.")
    print("Conclusion:", null_hypothesis)
```

The Kruskal-Wallis H-test is a non-parametric method used to test whether samples originate from the same distribution. It is used for comparing two or more independent samples of equal or different sample sizes.

the null hypothesis (H0) states that there is no significant difference in Average IQ among different Population Classifications in Asia. The test statistic is 1.51 and the p-value is approximately 0.68.

A p-value is a measure of the probability that an observed difference could have occurred just by random chance. The lower the p-value, the greater the statistical significance of the observed difference. P-value can be used in hypothesis testing to help you support or reject the null hypothesis. It represents the probability that the results of your test occurred at random. If p-value is small (typically ≤ 0.05), it is evidence that the null hypothesis can be rejected.

However, the p-value is not less than 0.05, which means we fail to reject the null hypothesis. This suggests that the differences in Average IQ among different Population Classifications in Asia could have happened by chance and are not statistically significant.

In conclusion, based on the Kruskal-Wallis H-test, there is no evidence to suggest that there is a significant difference in Average IQ among different Population Classifications in Asia. This means that the Average IQ is likely similar across different Population Classifications in Asia. However, it’s important to note that failing to reject the null hypothesis does not prove it true. It simply suggests that the data do not provide strong evidence against it. Other factors could also influence IQ, and further research might be needed to explore these possibilities.

```python
numerical_col = df.select_dtypes(include=['float64', 'int64']).columns
correlation = df[numerical_col].corr(method = "spearman")
matriks = np.triu(correlation)
display(correlation)

plt.figure(figsize = (15, 10))
sns.heatmap(
    data = correlation,
    annot = True,
    fmt = ".2f",
    cmap = "cividis",
    center = 0,
    mask = matriks
)

plt.title("Correlation Between Numerical IQ", size=16, pad=10, fontweight='bold')
plt.show()
```

```python

numeric_columns = ['Literacy Rate',
       'Nobel Prize', 'HDI (2021)', 'Mean years of schooling - 2021',
       'GNI_2021', 'Population - 2023', 'Average IQ']

# Select numeric columns from DataFrame
data = df[numeric_columns]

# Handling missing values by dropping rows with any missing values
data = data.dropna()

# Calculate correlation matrix
correlation_matrix = data.corr(method='spearman')

# Define function to assess correlation strength
def assess_correlation(correlation):
    kuatKorelasi_dict = {
        (0, 0.2): "sangat lemah (very weak)",
        (0.2, 0.4): "lemah (weak correlation)",
        (0.4, 0.6): "sedang (moderate correlation)",
        (0.6, 0.9): "kuat (strong correlation)",
        (0.9, 1): "sangat kuat (very strong correlation)"
    }

    result = []

    for col in correlation.columns:
        for i, kuat in kuatKorelasi_dict.items():
            corr_value = correlation.loc[col, 'Average IQ']
            if i[0] <= abs(corr_value) <= i[1]:
                kuatKorelasi = kuat
                break

        arahKorelasi = "positif" if corr_value > 0 else "negatif" if corr_value < 0 else "netral"

        result.append([col, round(corr_value, 5), kuatKorelasi, arahKorelasi])

    df_result = pd.DataFrame(result, columns=['Column', 'Correlation', 'Strength', 'Direction'])
    df_result.set_index('Column', inplace=True)

    return df_result

# Test the function
result_df = assess_correlation(correlation_matrix)

print('\033[4mCORRELATION SUMMARY BY AVERAGE IQ\033[0m')
result_df
```

The data  shows the correlation between various factors and the Average IQ. Here are some insights based on the correlation coefficients:

This suggests that as these factors increase, the Average IQ also tends to increase. These factors are all related to education and socio-economic conditions, which are known to influence IQ.

3. Population - 2023 (-0.02738) has a very weak negative correlation with Average IQ. This suggests that there is almost no relationship between the population size and the Average IQ.
